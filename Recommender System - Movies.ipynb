```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("C:\\Users\\Chris\\Desktop\\Python Learning\\DataSets\\ratings_Beauty.csv")
#Trimming the datset as it is long for processing
mov = df.tail(25000)
print(mov.shape)
mov.head()
```

```python
#Movie name against which recommendation has to be projected
movie_name = 'B00IBS9QC6'
```

```python
#Checking whether the data has any null values for data cleaning purposes
mov.isnull().sum()
```

```python
#Checking for variables types and its numbers and stats to understand the data
mov.describe(include = 'all')
```

```python
# Seperating the column and  sorting by unique combinations of movie name and getting the count
popular = mov.groupby('ProductId').count()
pop = pd.DataFrame(popular['Rating'])
print(pop.shape)
pop.head(2)
```

```python
# Sepearting the columns and sorting by movie name and rating with mean to get the average rating for each movie
most_pop_rate = mov.groupby('ProductId')['Rating'].mean()
print(most_pop_rate.shape)
most_pop_rate.head(5)
```

```python
# adding count and average rating to one dataframe 
pop['Avg of Rating'] = most_pop_rate
pop.head(5)
```

```python
#sort by rating with highest count of rating
pop.sort_values('Rating', ascending = False).head(10)
```

```python
#Creating the pivot for calculating correleation
rate = mov.pivot_table(values = 'Rating', index = 'UserId', columns = 'ProductId')
rate
```

```python
# Against movie name list of users showing what rating they have provided
rate[movie_name]
```

```python
# Calculating correlation against movie names by using rating and userid
rate.corrwith(rate[movie_name]).dropna()
```

```python
# Creating datafrome with the correlated contents and dropping null values
corr_movie = pd.DataFrame(rate.corrwith(rate[movie_name]), columns = ['Corr'])
corr_movie.dropna(inplace = True)
corr_movie.head(5)
```

```python
#Joiting Correlation and count of rating to get the best recommendations.
corr_movie = corr_movie.join(pop['Rating'])
```

```python
corr_movie
```

```python
# Selecting the movie which has highest correleation based on highest count of rating
Recommendation = corr_movie[corr_movie['Rating']>10].sort_values('Corr', ascending = False)
print(Recommendation.shape)
Recommendation
```

```python
#Recommended Movies
l= list(Recommendation.index)
l.remove(movie_name)
for x in l[0:10]:
    print(x)
```

```python
#Random movie selection
movie_name = list(mov['ProductId'].sample())[0]
movie_name
```

```python
popular = mov.groupby('ProductId').count()
pop = pd.DataFrame(popular['Rating'])
most_pop_rate = mov.groupby('ProductId')['Rating'].mean()
pop['Count of Rating'] = most_pop_rate
pop.sort_values('Rating', ascending = False).head(10)
rate = mov.pivot_table(values = 'Rating', index = 'UserId', columns = 'ProductId')
corr_movie = pd.DataFrame(rate.corrwith(rate[movie_name]), columns = ['Corr'])
corr_movie.dropna(inplace = True)
corr_movie = corr_movie.join(pop['Rating'])
Recommendation = corr_movie[corr_movie['Rating']>10].sort_values('Corr', ascending = False)
l= list(Recommendation.index)
l.remove(movie_name)
for x in l[0:10]:
    print(x)
```
